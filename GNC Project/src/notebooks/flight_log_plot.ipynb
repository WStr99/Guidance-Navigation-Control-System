```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/flight_log_detailed.csv")

plt.figure(figsize=(16, 4), dpi=150)  # width=16 inches, height=6 inches
plt.plot(df["Time"], df["xTrue"], label="x True")
plt.plot(df["Time"], df["xEstimate"], label="x Estimate")
plt.legend()
plt.title("X-Axis Position Tracking")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.grid(True)
plt.show()

plt.figure(figsize=(16, 4), dpi=150)  # width=16 inches, height=6 inches
plt.plot(df["Time"], df["yTrue"], label="y True")
plt.plot(df["Time"], df["yEstimate"], label="y Estimate")
plt.legend()
plt.title("Y-Axis Position Tracking")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.grid(True)
plt.show()

plt.figure(figsize=(16, 4), dpi=150)  # width=16 inches, height=6 inches
plt.plot(df["Time"], df["zTrue"], label="z True")
plt.plot(df["Time"], df["zEstimate"], label="z Estimate")
plt.legend()
plt.title("Z-Axis Position Tracking")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.grid(True)
plt.show()
```

```python
# Create a figure with two 3D subplots
fig = plt.figure(figsize=(18, 8), dpi=150)

# First viewpoint
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot(df["xTrue"], df["yTrue"], df["zTrue"], label='True Position', linewidth=2)
ax1.plot(df["xEstimate"], df["yEstimate"], df["zEstimate"], label='Estimated Position', linestyle='--', linewidth=2)
ax1.set_title("Viewpoint 1 (elev=0, azim=45)")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Z")
ax1.view_init(elev=0, azim=45)
ax1.legend()

# Second viewpoint
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot(df["xTrue"], df["yTrue"], df["zTrue"], label='True Position', linewidth=2)
ax2.plot(df["xEstimate"], df["yEstimate"], df["zEstimate"], label='Estimated Position', linestyle='--', linewidth=2)
ax2.set_title("Viewpoint 2 (elev=0, azim=120)")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_zlabel("Z")
ax2.view_init(elev=0, azim=120)
ax2.legend()

plt.tight_layout()
plt.show()
```
